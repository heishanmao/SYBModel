```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
```

```python
Years = pd.read_csv('./GCAM_full/20220421_gcam_production.csv')
Years = Years.columns.values.tolist()[5:-1] ## string

root = './Exps-tax/'
```

```python
data = pd.DataFrame()
for year in Years:
    path = root + 'SSP2/all_rates_'+ str(year) + '.csv'
    data0 = pd.read_csv(path)
    data = pd.concat([data,data0])
data
```

```python
def plot(ax, data):

    ax.bar(data['Year'], data['OBJ_Values']/ 1e15)
    #ax.bar(data['Year'], data['Revenue_Total'] / 1e15)
    #ax.bar(data['Year'], data['Cost_Total'] / 1e12)


data0 = data.query('Tax == 0')
data1 = data.query('Tax == 1')

fig = plt.figure(figsize=(25, 10),facecolor='#faf8ed')
#fig.tight_layout()
gs = GridSpec(2, 3, hspace=0.0, wspace=0.2, width_ratios=[1,1,1])

ax = plt.subplot(gs[0, 0])
plot(ax, data0)

ax = plt.subplot(gs[0, 1])
plot(ax, data1)
```
